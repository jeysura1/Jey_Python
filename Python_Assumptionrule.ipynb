```python
groc_series  = pd.DataFrame(pd.Series(groc_list))
groc_series = groc_series.iloc[:9835,:] # removing the last empty transaction
groc_series
groc_series.columns = ["transactions"]
groc_series
```

```python
X = groc_series['transactions'].str.join(sep='*').str.get_dummies(sep='*')
X
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)
```
