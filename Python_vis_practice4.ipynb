```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
data=pd.read_csv("train.csv")
```

```python
data.head()
```

```python
(data.dtypes)
```

```python
data.groupby('Survived')['PassengerId'].count()
```

```python
#Visual Exploration
f, ax=plt.subplots(figsize=(11,5)) # f fig and ax axes
sns.boxplot(x='Survived',y='Age', data=data);
```

```python
f, ax=plt.subplots(figsize=(11,5)) # f fig and ax axes
sns.barplot(y='Survived',x='Sex', data=data);
```

```python
f, ax=plt.subplots(figsize=(11,5)) # f fig and ax axes
sns.boxplot(x='Sex',y='Age', hue='Survived', data=data);
```

```python
sns.barplot(x='Pclass',y='Survived', data=data);
```

```python
sns.barplot(x='Pclass',y='Survived',hue='Sex', data=data);
```

```python
sns.barplot(x='SibSp',y='Survived', data=data);
```

```python
sns.barplot(x='Parch',y='Survived', data=data);
```

```python
survived= data.loc[data['Survived']==1, 'Age'].dropna() # Age of survived ones
sns.distplot(survived)
plt.title('Survived')
```

```python
not_survived= data.loc[data['Survived']==0, 'Age'].dropna() # Age of Not survived ones
sns.distplot(not_survived)
plt.title('Not Survived')
```

```python
sns.pairplot(data.dropna());
```

```python
# kind of pivot between survived with class find out age counts
grid=sns.FacetGrid(data, col='Survived', row='Pclass', size==2.4, aspect=1.5)
grid.map(plt.hist,'Age', alpha=.5, bins=20)
grid.add_legend()
```
