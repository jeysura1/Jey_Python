## Imputation

```python
#Hot-Deck
# looking at the data before imputation
df_hot_deck = df.copy()
df[887:]
```

```python
# looking at the data after imputation
df_hot_deck.fillna(method="ffill", inplace=True)
df_hot_deck[887:]
```

```python
#Statistical method
# View all rows with a missing age
df3 = df.copy()
missing_age = df3[df3["Age"].isna()].index
df3[df3["Age"].isna()]
```

```python
missing_age
```

```python
# imputing all rows with missing age with the mean age
df3["Age"].fillna(df3["Age"].mean(), inplace=True)
df3.loc[missing_age]
```

```python
# current value counts of the embarked column
df3["Embarked"].value_counts(dropna=False)
```

```python
# view rows with missing values
missing_embarked = df3[df3["Embarked"].isna()].index
df3[df3["Embarked"].isna()]
```

```python
# fill the rows with missing values and view them with mode
df3["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
df3.loc[missing_embarked]
```

```python
# KNN Imputation
from sklearn.impute import KNNImputer
X = df[["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]].copy() 
X
```

```python
knn_imputer = KNNImputer(n_neighbors=3)
filled_df = pd.DataFrame(data=knn_imputer.fit_transform(X),
                         columns=["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"])
filled_df.loc[missing_age]
```

KNNImputer states "Each samples missing values are imputed using the mean value from n_neighbors nearest neighbors found in the training set. Two samples are close if the features that neither is missing are close."
