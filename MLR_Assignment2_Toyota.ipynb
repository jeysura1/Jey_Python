## VIF value validation

```python
rsq_Age= smf.ols('Age_08_04~KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=car1).fit().rsquared  
vif_Age = 1/(1-rsq_Age)

rsq_km = smf.ols('KM~Age_08_04+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=car1).fit().rsquared  
vif_km = 1/(1-rsq_km)

rsq_hp = smf.ols('HP~Age_08_04+KM+cc+Doors+Gears+Quarterly_Tax+Weight',data=car1).fit().rsquared  
vif_hp = 1/(1-rsq_hp) 

rsq_cc = smf.ols('cc~Age_08_04+KM+HP+Doors+Gears+Quarterly_Tax+Weight',data=car1).fit().rsquared  
vif_cc = 1/(1-rsq_cc)

rsq_doors = smf.ols('Doors~Age_08_04+KM+HP+cc+Gears+Quarterly_Tax+Weight',data=car1).fit().rsquared  
vif_doors = 1/(1-rsq_doors)

rsq_gear = smf.ols('Gears~Age_08_04+KM+HP+cc+Doors+Quarterly_Tax+Weight',data=car1).fit().rsquared  
vif_gear = 1/(1-rsq_gear) 

rsq_wt = smf.ols('Weight~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax',data=car1).fit().rsquared  
vif_wt = 1/(1-rsq_wt) 

rsq_tax = smf.ols('Quarterly_Tax~Age_08_04+KM+HP+cc+Doors+Gears+Weight',data=car1).fit().rsquared  
vif_tax = 1/(1-rsq_tax) 

# Storing vif values in a data frame
vif = {'Variables':['age','km','hp','cc','doors','gear','wt','tax'],'VIF':[vif_Age,vif_km,vif_hp,vif_cc,vif_doors,vif_gear,vif_wt,vif_tax]}
Vif_frame = pd.DataFrame(vif)  
Vif_frame
```

## Model validation

```python
#Residual Vs Regressors
yfig=plt.figure(figsize=(10,7))
fig=sm.graphics.plot_partregress_grid(model4,fig=fig)
plt.show()
```

```python
#Test for Normality of Residuals (Q-Q Plot)
import statsmodels.api as sm
qqplot=sm.qqplot(model4.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
#Residual Plot for Homoscedasticity
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model4.fittedvalues),
            get_standardized_values(model4.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

## Not finding any pattern, so the model is good
