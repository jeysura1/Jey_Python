```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set_style ("darkgrid", {'axes.grid' : True})
sns.lmplot(x='total_bill', y='tip', data=tips) # linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, hue='sex', palette='afmhot')
plt.show() # Set is color pattern
```

```python
#Univariable: strip plot
sns.stripplot(y= 'tip', data=tips, jitter=True)
plt.ylabel('tip =($)')
plt.show()
#Jitter is density
```

```python
tips.head()
```

```python
sns.stripplot(x='day', y= 'tip', data=tips, size=4, jitter=True)
plt.ylabel('tip =($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x='day', y= 'tip', data=tips)
plt.ylabel('tip =($)')
plt.show()
```

```python
sns.swarmplot(x='day', y= 'tip', data=tips, hue='sex')
plt.ylabel('tip =($)')
plt.show()
```

```python
sns.swarmplot(x='tip', y= 'day', data=tips, hue='sex',orient='h')
plt.ylabel('tip =($)')
plt.show()
```

```python
#plot subplot
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')
```

```python
#violin plot
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#subplot
plt.subplot(1,2,1) # size and position of the plot
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#combining plots
sns.violinplot(x='day',y='tip',data=tips, inner=None, color='yellow')
sns.stripplot(x='day', y= 'tip', data=tips, size=4, jitter=True)
plt.ylabel('tip =($)')
```

```python
#Bivariate >"two variables" data visualizations
#joint plot
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
#Density  plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```python
#pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```
